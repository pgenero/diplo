# **EVALUAR MODELO LINEAL 2015-2020**

## Probar el modelo lineal

> Voy a tratar de ver si los resultados de cada año agrupados por mes se ajustan a un modelo lineal. Entonces aplico los diferentes tests del modelo de regresión lineal simple.

```r
lm1520=lm(resultado/1000000~mesesf-1)
summary(lm1520)
```

> Test Shapiro-Wilk - el p-value < 0.05 → los residuos no son normales

```r
shapiro.test(rstandard(lm1520))
```

```r
shapiro.test(residuals(lm1520))
```

> Test Breusch-Pagan - el p-value < 0.05 → las varianzas no son iguales

```r
bptest(lm1520) 
```

> Test Durbin-Watson - el p-value < 0.05 → los residuos no son independientes

```r
dwtest(lm1520)
```

```r
plot(lm1520)
```

*De acuerdo a lo que pude ir leyendo acerca del estudio de las series de tiempo, se recomienda realizar alguna transformación para estabilizar la serie antes de estimar los modelos, sobre todo si se pretende hacer uso de ARIMA, que requiere series estacionarias. Por lo que vi en las gráficas anteriores esta serie carece de estacionariedad.*

*Lo que buscan estas transformaciones es una media y varianza constantes a lo largo de la serie, para que se verifique la estacionariedad.*

*Inicialmente voy a probar con una diferenciación simple de  serie, y luego esos datos diferenciados, someterlos a una transformación.*

## Diferenciar variable Resultado Neto

> Diferenciación de la serie de tiempo de datos 2015 a 2020

```r
diff1520=diff(rdo.meses)
diff1520
```

```r
 diffmesesf=mesesf[-1] 
 #Acá le saco un el primer mes a mi objeto "mesesf", ya que al diferenciar el primer valor que es Julio pierde el dato (queda en 0)
 # De esta forma puedo hacer luego el test lm
```

```r
lm_diff=lm(diff1520~diffmesesf-1)
summary(lm_diff)
```

> Test Shapiro-Wilk - el p-value < 0.05 → los residuos no son normales

```r
shapiro.test(rstandard(lm_diff))
```

```r
shapiro.test(residuals(lm_diff))
```

> Test Breusch-Pagan - el p-value < 0.05 → las varianzas no son iguales

```r
bptest(lm_diff) 
```

> Test Durbin-Watson - el p-value > 0.05 → los residuos son independientes

```r
dwtest(lm_diff) #Logra la independencia de residuos
```

```r
plot(lm_diff)
```

```r
auto.arima(diff1520,seasonal = T,ic = c("aicc"), stepwise=T, allowdrift = T) #no son independientes los errores? es MA(3)
```

*Como sólo logre probar la independencia de los residuos, decido probar con transformar la serie diferenciada. Aplico la transformación de Johnson.*

## Transformación de Johnson

```r
trdiff1520=RE.Johnson(diff1520)
```

```r
trdiff1520_lm=lm(trdiff1520$transformed~diffmesesf-1)
summary(trdiff1520_lm)
```

> Test Shapiro-Wilk - el p-value > 0.05 → los residuos son normales

```r
shapiro.test(rstandard(trdiff1520_lm))
```

```r
shapiro.test(residuals(trdiff1520_lm)) #Se logra normalidad en los residuos
```

> Test Breusch-Pagan - el p-value < 0.05 → las varianzas no son iguales

```r
bptest(trdiff1520_lm)
```

> Test Durbin-Watson - el p-value > 0.05 → los residuos son independientes

```r
dwtest(trdiff1520_lm) #Logra la independencia de residuos
```

```r
trdiff1520=ts(trdiff1520$transformed, start = c(2015,8), frequency = 12)
# La variable transformada, la llevo a una serie de tiempo, para aplicarle la función auto.arima
```

```r
auto.arima(trdiff1520,seasonal = T,ic = c("aicc"), stepwise=T, allowdrift = T)
```

```r
plot(trdiff1520_lm)
```

*Luego de la transformación, se pudo lograr la normalidad en los residuos (ver el segundo gráfico) y la independencia, pero no así la homocedasticidad.*
